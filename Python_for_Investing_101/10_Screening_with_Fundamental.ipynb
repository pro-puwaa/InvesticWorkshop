## Import library

```python
import yfinance as yf
import mplfinance as mpf
import pandas as pd
import numpy as np
import pandas_ta as ta
```

## Import data

```python
path = 'https://drive.google.com/uc?export=download&id=1EqhMOcAPX7WfW1gObCPPsFq99D5RfA42'
```

```python
raw_df = pd.read_excel(path)
raw_df
```

```python
df = raw_df.copy()
```

```python
df.dtypes
```

## Data Cleaning

```python
df = df.replace('-', np.nan)

df['Yield%'] = df['Yield%'].astype(float)
df['P/BV'] = df['P/BV'].astype(float)
df['DPS'] = df['DPS'].astype(float)

df['P/E'] = round(df['P/E'], 1)
```

```python
df.dtypes
```

```python
df
```

## Screening Assets

```python
df[(df['P/E'] > 15) & (df['P/E'] < 80)].sort_values(by='P/E', ascending=False)
```

```python
df[(df['P/E'] > 15) & (df['P/E'] < 80)].sort_values(by=['P/E', 'MCap (M)'], ascending=[True, False]).head(25)[['Name', 'P/E', 'MCap (M)']]
```
