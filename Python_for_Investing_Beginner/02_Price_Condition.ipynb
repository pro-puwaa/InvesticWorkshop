## Import Libraries

```python
import pandas as pd
import numpy as np
import yfinance as yf
```

```python
raw_df = yf.download('BTC-USD', start='2021-01-01', end='2022-05-01')
raw_df
```

## Filter close price > 60k USD

```python
raw_df[raw_df['Close'] > 60000].head(10)
```

## Filter return > 0

```python
raw_df['return'] = raw_df.Close.pct_change()
raw_df
```

```python
raw_df[raw_df['return'] > 0]
```

## Create function

สร้าง Function แล้วให้ Print โดยมี Condition ที่ว่าหาก Return > 0 ให้ Print “Gain” หาก Return < 0 ให้ปริ้น "Loss"

```python
def gain_loss_return(df):
    for i in range(len(raw_df)):
        # "-1" = return column
        if(df.iloc[i, -1] > 0):
            print('Gain')
        elif (df.iloc[i, -1] < 0):
            print('Loss')
```

```python
gain_loss_return(raw_df)
```
